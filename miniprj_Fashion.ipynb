```python
X = mnist_fashion['data']
y = mnist_fashion['target']

X.shape, y.shape
```

```python
classes = ['tshirt', 'trouser', 'pullover', 'dress','coat',
            'sandal', 'shirt', 'sneaker', 'bag', 'ankle boot']
```

```python
# visualize with 2x5 grid
plt.rcParams['figure.figsize'] = (5,5)  #high quality fig default size(5,5)
plt.rcParams['figure.dpi'] = 100

digits = y.unique()  #plot classes
digits = [int(x) for x in digits]  #list comprehension 
digits.sort()
digits
```

```python
plt.rcParams['figure.figsize'] = (15,4)
plt.rcParams['figure.dpi'] = 200

for i in digits:
  plt.subplot(2,5, i+1)  #2r 5c
  plt.imshow(X[y==str(i)].iloc[0].to_numpy().reshape(28,28))
  plt.title(classes[i])

plt.tight_layout()
```

```python
#Designing neural networks
#split dataset
from sklearn.model_selection import train_test_split
```

```python
y = y.astype('int')
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X/255, y, test_size=0.2, random_state=42,
                                                    stratify=y)

X_train.shape, X_test.shape
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Sequential 
```

```python
#building neural network model
def get_model():
  model = Sequential()
  model.add(Flatten(input_shape=(X.shape[1],)))  
  model.add(Dense(64, activation='relu'))
  model.add(Dense(128, activation='relu'))
  model.add(Dense(10, activation='softmax')) #not binary
  return model
```

```python
model = get_model()
```

```python
model.summary()
```

```python
784*64+64, 64*128+128, 128*10+10
```

```python
model.layers
```

```python
model.layers[2].name
```

```python
weight1 =model.layers[1].get_weights()
```

```python
weight1[0].shape, weight1[1].shape
```

```python
#complilation
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
#training dataset
model.fit(X_train, y_train, epochs=10)
model.evaluate(X_test, y_test)
```

```python
plt.rcParams['figure.figsize'] = (15,4)
plt.rcParams['figure.dpi'] = 200

for i in digits:
  plt.subplot(2,5, i+1)
  test = X_test[y==i].iloc[0].to_numpy()
  plt.imshow(test.reshape(28,28))

  pred = model.predict(test.reshape(1,784,))
  idx = np.argmax(pred[0])
  pred_class = classes[idx]

  plt.title('original: {} | pred: {}'.format(classes[i], pred_class))

plt.tight_layout()
```
