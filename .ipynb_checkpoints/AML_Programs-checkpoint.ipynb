```python
#practical no.1 kmeans

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import sklearn
```

```python
dataset = pd.read_csv(r'E:\Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

```python
from sklearn.cluster import KMeans
wcss = [] 
for i in range(1, 11): 
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X) 
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = "k-means++", random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
# Plotting the clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=60, c='red', label='Cluster1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=60, c='blue', label='Cluster2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=60, c='green', label='Cluster3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=60, c='violet', label='Cluster4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s=60, c='yellow', label='Cluster5')

# Plotting the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='black', label='Centroids')

# Adding labels and legend
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()

# Display the plot
plt.show()
```

```python
list_of_transactions = []
#Append the list
for i in range(0, 7501):
    list_of_transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
list_of_transactions[0]
```

```python
from apyori import apriori
rules = apriori(list_of_transactions, min_support = 0.004, min_confidence = 0.2, min_lift = 3, min_length = 2)
# Create a list of rules and print the results
results = list(rules)
#Here is the first rule in list or results
results[0]
```

```python
def inspect(results):
    lhs     =  [tuple(result [2] [0] [0]) [0] for result in results]
    rhs     =  [tuple(result [2] [0] [1]) [0] for result in results]
    supports = [result [1] for result in results]
    confidences = [result [2] [0] [2]   for result in results]
    lifts = [result [2] [0] [3]   for result in results]
    return list(zip(lhs,rhs,supports,confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results),columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'] )
resultsinDataFrame.head(10)
```

```python
#practical no. 5 reinforecment 1-Calculating rewards

import random
from typing import List
class SampleEnviornment:
  def __init__(self):
    self.steps_left = 20
  def get_observation(self):
    return [0.0, 0.0, 0.0]
  def get_action(self):
    return [-1, 1]
  def is_done(self):
    return self.steps_left <= 0
  def action(self, action:int):
    if self.is_done():
      raise Exception("Game is Over")
    self.steps_left -= 1
    return random.random()
class Agent:
  def __init__(self):
    self.total_reward = 0
  def step(self, env):
    current_obs = env.get_observation()
    print("Observation{}".format(current_obs))
    actions = env.get_action()
    print(actions)
    reward = env.action(random.choice(actions))
    self.total_reward += reward
if __name__ == '__main__':
  env = SampleEnviornment()
  agent = Agent()
  i=0
  while not env.is_done():
    i+=1
    print("Step-", i)
    agent.step(env)
    print("Total reward got: %.4f"%agent.total_reward)
```

```python
#2-discount rewards

def calculate_discounted_reward(rewards, gamma):
    discounted_reward = 0.0
    cumulative_reward = 0.0
    for t in reversed(range(len(rewards))):
        cumulative_reward = rewards[t] + gamma * cumulative_reward
        discounted_reward += cumulative_reward
    return discounted_reward

# Example list of rewards
rewards = [1, 2, 3, 4, 5]
# Example discount factor
gamma = 0.9

total_discounted_reward = calculate_discounted_reward(rewards, gamma)
print(f'Total discounted reward: {total_discounted_reward}')
```

```python
#3-calculating optimal quantities

import numpy as np

def value_iteration(states, actions, transition_probabilities, rewards, gamma, theta):
    V = {state: 0 for state in states} 
    policy = {state: None for state in states} 
    
    while True:
        delta = 0 
        for s in states:
            v = V[s] 
            action_values = {}
            for a in actions:
                action_value = sum(transition_probabilities[s][a][s_prime] * 
                                   (rewards[s][a][s_prime] + gamma * V[s_prime])
                                   for s_prime in transition_probabilities[s][a])
                action_values[a] = action_value
            best_action_value = max(action_values.values())
            V[s] = best_action_value
            delta = max(delta, abs(v - V[s])) 
        if delta < theta:
            break 

    for s in states:
        action_values = {}
        for a in actions:
            action_value = sum(transition_probabilities[s][a][s_prime] * 
                               (rewards[s][a][s_prime] + gamma * V[s_prime])
                               for s_prime in transition_probabilities[s][a])
            action_values[a] = action_value
        best_action = max(action_values, key=action_values.get)
        policy[s] = best_action
    
    return V, policy

# Define the MDP
states = ['s1', 's2']
actions = ['a1', 'a2']
transition_probabilities = {
    's1': {'a1': {'s1': 0.7, 's2': 0.3}, 'a2': {'s1': 0.4, 's2': 0.6}},
    's2': {'a1': {'s1': 0.8, 's2': 0.2}, 'a2': {'s1': 0.5, 's2': 0.5}}
}
rewards = {
    's1': {'a1': {'s1': 5, 's2': 10}, 'a2': {'s1': -1, 's2': 2}},
    's2': {'a1': {'s1': 1, 's2': 3}, 'a2': {'s1': 0, 's2': -1}}
}
gamma = 0.9
theta = 0.0001

# Run value iteration
optimal_values, optimal_policy = value_iteration(states, actions, transition_probabilities, rewards, gamma, theta)

# Print the results
print("Optimal Values:")
for state, value in optimal_values.items():
    print(f"Value of {state}: {value:.2f}")

print("\nOptimal Policy:")
for state, action in optimal_policy.items():
    print(f"Best action for {state}: {action}")
```

```python
#4] Implementing Q Learning:

import numpy as np
import random

class QLearningAgent:
    def __init__(self, states, actions, alpha=0.1, gamma=0.9, epsilon=0.1):
        self.states = states
        self.actions = actions
        self.alpha = alpha
        self.gamma = gamma
        self.epsilon = epsilon
        self.q_table = {state: {action: 0.0 for action in actions} for state in states}
    
    def choose_action(self, state):
        if random.uniform(0, 1) < self.epsilon:
            return random.choice(self.actions)
        else:
            return max(self.q_table[state], key=self.q_table[state].get)
    
    def update_q_value(self, state, action, reward, next_state):
        best_next_action = max(self.q_table[next_state], key=self.q_table[next_state].get)
        td_target = reward + self.gamma * self.q_table[next_state][best_next_action]
        td_error = td_target - self.q_table[state][action]
        self.q_table[state][action] += self.alpha * td_error
    
    def train(self, episodes, max_steps_per_episode, environment):
        for episode in range(episodes):
            state = environment.reset()
            for step in range(max_steps_per_episode):
                action = self.choose_action(state)
                next_state, reward, done = environment.step(state, action)
                self.update_q_value(state, action, reward, next_state)
                state = next_state
                if done:
                    break
    
    def get_policy(self):
        policy = {}
        for state in self.states:
            policy[state] = max(self.q_table[state], key=self.q_table[state].get)
        return policy

class ExampleEnvironment:
    def __init__(self):
        self.state_space = ['s1', 's2']
        self.action_space = ['a1', 'a2']
        self.transitions = {
            's1': {'a1': ('s2', 10, False), 'a2': ('s1', -1, False)},
            's2': {'a1': ('s1', -1, False), 'a2': ('s2', 1, False)}
        }
    
    def reset(self):
        return random.choice(self.state_space)
    
    def step(self, state, action):
        next_state, reward, done = self.transitions[state][action]
        return next_state, reward, done

states = ['s1', 's2']
actions = ['a1', 'a2']
environment = ExampleEnvironment()
agent = QLearningAgent(states, actions)
agent.train(episodes=1000, max_steps_per_episode=100, environment=environment)

optimal_policy = agent.get_policy()
print("Optimal Q-Values:")
for state in agent.q_table:
    print(f"{state}: {agent.q_table[state]}")
```

```python
#5] Setting up an Optimal Action:

agent.train(episodes=1000, max_steps_per_episode=100, environment=environment)
optimal_policy = agent.get_policy()
print("\nOptimal Policy:")
for state in optimal_policy:
    print(f"Best action for {state}: {optimal_policy[state]}")
```

```python
#practical no.6 -1. Checking Stationary

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
time_series = np.random.randn(100) + np.arange(100)

plt.figure(figsize=(10, 6))
plt.plot(time_series)
plt.title('Example Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()

def adf_test(time_series):
    result = adfuller(time_series)
    print('ADF Statistic: {:.4f}'.format(result[0]))
    print('p-value: {:.4f}'.format(result[1]))
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t{}: {:.4f}'.format(key, value))
    
    if result[1] < 0.05:
        print("Reject the null hypothesis; Data is stationary.")
    else:
        print("Fail to reject the null hypothesis; Data is non-stationary.")

print("\nAugmented Dickey-Fuller Test Results:")
adf_test(time_series)
```

```python
#2. Converting a non-stationary data to stationary

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate example time series data
np.random.seed(0)
time_series = np.random.randn(100) + np.arange(100)

# Plot the original time series data
plt.figure(figsize=(10, 6))
plt.plot(time_series)
plt.title('Original Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Define the function to difference the time series
def difference_series(data, interval=1):
    diff = []
    for i in range(interval, len(data)):
        value = data[i] - data[i - interval]
        diff.append(value)
    return pd.Series(diff)

# Apply differencing and plot the differenced time series
differenced_series = difference_series(time_series)
plt.figure(figsize=(10, 6))
plt.plot(differenced_series)
plt.title('Differenced Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Define the function to perform seasonal differencing
def seasonal_difference(data, period):
    diff = []
    for i in range(period, len(data)):
        value = data[i] - data[i - period]
        diff.append(value)
    return pd.Series(diff)

# Apply seasonal differencing and plot the seasonal differenced time series
seasonal_differenced_series = seasonal_difference(time_series, period=12)
plt.figure(figsize=(10, 6))
plt.plot(seasonal_differenced_series)
plt.title('Seasonal Differenced Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
#3. Implementing Dickey Fuller Test
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
np.random.seed(0)
time_series = np.random.randn(100) + np.arange(100)
plt.figure(figsize=(10, 6))
plt.plot(time_series)
plt.title('Original Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()
def adf_test(time_series):
    result = adfuller(time_series)
    print('ADF Statistic: {:.4f}'.format(result[0]))
    print('p-value: {:.4f}'.format(result[1]))
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t{}: {:.4f}'.format(key, value))
        if result[1] < 0.05:
            print("Reject the null hypothesis; Data is stationary.")
        else:
            print("Fail to reject the null hypothesis; Data is non-stationary.")
print("\nAugmented Dickey-Fuller Test Results:")
adf_test(time_series)
```

```python
#4. Time Series analysis to Plot ACF and PACF (Autocorrelation function and Partial Autocorrelation Function)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
np.random.seed(0)
time_series = np.random.randn(100) + np.arange(100)
plt.figure(figsize=(10, 6))
plt.plot(time_series)
plt.title('Original Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()
plt.figure(figsize=(10, 6))
plot_acf(time_series, lags=30)
plt.title('Autocorrelation Function (ACF)')
plt.xlabel('Lag')
plt.ylabel('ACF')
plt.show()
plt.figure(figsize=(10, 6))
plot_pacf(time_series, lags=30)
plt.title('Partial Autocorrelation Function (PACF)')
plt.xlabel('Lag')
plt.ylabel('PACF')
plt.show()
```

```python
#5. Time series Analysis to Generating the ARIMA plot (autoregressive integrated moving average)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.arima_process import ArmaProcess
np.random.seed(0)
arparams = np.array([0.5])
maparams = np.array([0.2])
ar = np.r_[1, -arparams] # add zero-lag and negate
ma = np.r_[1, maparams] # add zero-lag
arma_process = ArmaProcess(ar, ma)
time_series = arma_process.generate_sample(100)
plt.figure(figsize=(10, 6))
plt.plot(time_series)
plt.title('Original Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()
model = ARIMA(time_series, order=(1, 0, 1)) # Example: ARIMA(1,0,1)
fit_model = model.fit()
forecast = fit_model.forecast(steps=10) # Forecast 10 steps ahead
plt.figure(figsize=(10, 6))
plt.plot(time_series, label='Observed')
plt.plot(np.arange(len(time_series), len(time_series) + len(forecast)), forecast, 
color='red', label='Forecast')
plt.title('ARIMA Forecast')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#6. Time Series Analysis for TSA Forecasting

import pandas as pd
import numpy as np
from prophet import Prophet
import matplotlib.pyplot as plt

# Ensure reproducibility
np.random.seed(0)

# Create a time series data frame
dates = pd.date_range(start='2022-01-01', periods=100)
values = np.random.randn(100)
df = pd.DataFrame({'ds': dates, 'y': values})

# Initialize and fit the Prophet model
model = Prophet()
model.fit(df)

# Make future dataframe for forecasting
future = model.make_future_dataframe(periods=10, freq='D') # Forecasting 10 days ahead

# Predict future values
forecast = model.predict(future)

# Plot the forecast
fig = model.plot(forecast)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Prophet Forecast')
plt.grid(True)
plt.show()
```

```python
#practical no.7 IMPLIMENTATION OF BOOSTING: A. Cross validation:

from sklearn .model_selection import cross_val_score, KFold 
from sklearn.svm import SVC
from sklearn.datasets import load_iris 
iris = load_iris() 
X, y = iris.data, iris.target 
svm_classifier = SVC(kernel='linear') 
num_folds = 5 
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42) 
cross_val_results = cross_val_score(svm_classifier, X, y, cv=kf) 
print (f'Cross-Validation Results (Accuracy): {cross_val_results}') 
print (f'Mean Accuracy: {cross_val_results.mean()}') 
```
